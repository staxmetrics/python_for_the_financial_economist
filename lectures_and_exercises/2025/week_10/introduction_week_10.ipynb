# Week 11:  Fully-Flexible Views and CVaR optimization

In this week, we will look at the Fully-Flexible Views approach to portfolio optimization. For the interested, on the often applied alternative Black-Litterman approach is presented in a notebook..

## Lectures and exercises

The **lectures** cover

* Implementation of the method in [Attilio Meucci (2010), Fully Flexible Views: Theory and Practice](https://papers.ssrn.com/sol3/papers.cfm?abstract_id=1213325)
* CVaR optimization and specification as a convex optimization problem

The **exercises** cover

* Implementation of Black-Litterman
* Scenario analysis using Fully-Flexible Views
* Portfolio optimization using Fully-Flexible Views
